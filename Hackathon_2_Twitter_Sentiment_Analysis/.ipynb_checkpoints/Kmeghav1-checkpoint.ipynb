```python
train_corpus1=[]
for i in range(0, 7274):
    rev=' '.join(re.sub("[\_\{\}\'RT]", " ",train_corpus[i]).split())
    train_corpus1.append(rev)
```

```python
CONTRACTIONS = {"Its":"It is", "m":"am","ll":"will","droid":"android","u":"you","ve":"have","re":"are"}
train_corpus2=[]
for i in range(0,7274):
    tweet = train_corpus1[i].replace("’","'")
    words = tweet.split()
    reformed = [CONTRACTIONS[word] if word in CONTRACTIONS else word for word in words]
    tweet = " ".join(reformed)
    train_corpus2.append(tweet)
    



```

```python
SMILEY = {":‑(":"sad", ":‑)":"smiley","(A)":"angel","(6)": "devil","(L)":"love","(U)":"brokenheart",":'(":"crying"}
train_corpus4=[]
for i in range(0,7274):
    
    words = train_corpus2[i].split()
    reformed = [SMILEY[word] if word in SMILEY else word for word in words]
    tweet = " ".join(reformed)
    train_corpus4.append(tweet)


```
